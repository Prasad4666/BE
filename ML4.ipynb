```python
x=2
lr=0.01
precision=0.00001
previous_step_size=1
iters=1
max_iterator=1000
gf=lambda x: (x+3)**2
```

```python
gd=[]
while precision <previous_step_size and iters <max_iterator:
    prev=x
    x=x - lr * gf(prev)
    previous_step_size=abs(x-prev)
    iters+=1
    print(f"Iterator : {iters} and Value : {x}")
    gd.append(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
